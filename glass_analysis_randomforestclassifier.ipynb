```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv("glass_data.csv")
```

```python
data.head()
```

```python
len(data)
```

```python
data = data.drop(['Id'], axis =1)
```

```python
#extract
inputs = data.iloc[:,:-1]
targets = data.iloc[:,-1]
```

```python
inputs.head()
```

```python
targets.head()
```

```python
train_input, test_input, train_target, test_target = train_test_split(inputs, targets, test_size =0.2, random_state=42)
```

```python
clf = RandomForestClassifier(random_state=26)
```

```python
clf.fit(train_input, train_target)
```

```python
clf.predict([train_input.iloc[125]])
```

Test

```python
test_pred = clf.predict(test_input)
```

```python
print(classification_report(test_target, test_pred))
```

### Summary  

A Random Forest Classifier is a powerful machine learning model, and to evaluate how well it performs, we use various metrics. When you get a detailed classification report, especially from a tool like scikit-learn in Python, you'll see F1-Score, Recall, Precision, and Support for each class.These metrics are essential for understanding the model's performance beyond simple accuracy, especially in situations where the data is imbalanced.  

#### Random Forest Evaluation Metrics 

These metrics are derived from the Confusion Matrix, which tabulates the model's correct and incorrect predictions.  

1. Precision
Definition: Precision answers the question: Out of all the instances the model predicted belonged to this class, how many actually did?$$\text{Precision} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Positives}}$$Goal: A high precision means the model is very reliable when it predicts that class; it rarely makes a "false alarm."Context: Important when the cost of a False Positive (wrongly predicting a positive) is high. Example: Filtering spam (you don't want a legitimate email classified as spam).

2. Recall (Sensitivity)

Definition: Recall answers the question: Out of all the instances that actually belonged to this class, how many did the model correctly identify?$$\text{Recall} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Negatives}}$$Goal: A high recall means the model misses very few actual instances of that class; it has a low chance of a "false negative."Context: Important when the cost of a False Negative (wrongly predicting a negative/missing an actual positive) is high. Example: Detecting a critical disease (you don't want to miss an actual case of the disease).

3. F1-Score

Definition: The F1-Score is the harmonic mean of Precision and Recall. It provides a single score that balances both metrics.$$\text{F1-Score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}$$Goal: Used when you need a balance between Precision and Recall. It is a more robust measure of overall model performance than accuracy, especially when dealing with imbalanced datasets (where one class has many more examples than others).Context: A good F1-score indicates the model is strong at both making correct positive predictions and capturing most of the actual positive cases.

4. Support

Definition: Support simply indicates the number of actual occurrences of the class in the specified dataset (usually the testing or validation set).Goal: To provide context for the other metrics.Context:If a class has a very low Support (e.g., 5 instances), the Precision and Recall metrics for that class might be unreliable or volatile.If a class has high Support (e.g., 1,000 instances), the metrics are generally more reliable.It helps you identify imbalance in your test data.  

## Summary for Random Forest  
When evaluating your Random Forest Classifier's output:Look at the F1-Score for the Minority Class: If your data is imbalanced (low support for one class), the F1-score for that smaller class is often the most critical metric for judging how well the model handles the difficult cases.High Precision + Low Recall: The model is highly accurate when it does predict the class, but it misses many actual instances.Low Precision + High Recall: The model captures almost all instances of the class, but it often includes false alarms (wrong predictions).
